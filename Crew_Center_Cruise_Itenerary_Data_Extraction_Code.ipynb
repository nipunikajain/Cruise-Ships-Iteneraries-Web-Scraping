```python
def iteration(x):
    var1 = "http://crew-center.com"
    var2 = "{}".format(x)
  
    url = "".join([var1, var2])
    print(url)

    html_content = requests.get(url).text
    soup = BeautifulSoup(html_content, "lxml")
    item = []
    df = pd.DataFrame()

    for table in soup.findAll('td',{"class": re.compile("xl")}):
        item.append(table.text)

    if len(item)==0:
        print('No Iteneary for {}'.format(x))
        
    
    elif item[3]=='Departure':
        
        N=4
        subList = [item[n:n+N] for n in range(0, len(item), N)]
        df1 = df.append(subList)

        header = df1.iloc[0] 
        df1 = df1[1:]
        df1.columns = header

        df1['Arrival-Departure'] = df1['Arrival'] +'-'+ df1['Departure']
        df1['ShipName'] = "{}".format(x)

        df1.drop(columns=['Arrival','Departure'],inplace=True)
        finalDF = pd.DataFrame()
        finalDF.append(df1)
        df1.to_csv(r'C:\Users\jainni1\OneDrive - Ecolab\Documents\My_Projects\Cruise_Order_Acquisition\itenerary\iter1.csv',
                   mode='a',header=False)
        
    else :
        N=3
        subList = [item[n:n+N] for n in range(0, len(item), N)]
        df1 = df.append(subList)
    
        header = df1.iloc[0] 
        df1 = df1[1:]
        df1.columns = header
        df1['ShipName'] = "{}".format(x)
#         print(df1)
        finalDF = pd.DataFrame()
        finalDF.append(df1)
        df1.to_csv(r'C:\Users\jainni1\OneDrive - Ecolab\Documents\My_Projects\Cruise_Order_Acquisition\itenerary\iter2.csv',
                   mode='a',header=False)
```
